```python
import pandas as pd
import os
```

# Removing outliers based on IQR 

```python
dftrain
```

```python
#taking 4th and 96th percentile instead of 25 and 75 as 25,75 pair was identifying in large percent of data as outlier
#we wanted to remove only extreme values
def remove_outlier_IQR(df):
    Q1=df.quantile(0.04)
    Q3=df.quantile(0.96)
    IQR=Q3-Q1
    df_final=df[~((df<(Q1-1.5*IQR)) | (df>(Q3+1.5*IQR)))]
    return df_final
```

```python
dftrain_nooutlier=remove_outlier_IQR(dftrain)
```
